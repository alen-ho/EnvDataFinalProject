**The model used in the final project is DeepLabv3+, a deep learning model with billions of parameters. Unlike traditional statistical models that rely on coefficients, DeepLabv3+ employs a complex network architecture. In this code, we provide an overview of the model's structure.**

**Model structure**

```python
import tensorflow as tf
from tensorflow.keras.models import load_model

# Load the model
model_path = "deeplabv3_model.h5"
model = load_model(model_path, custom_objects={'MeanIoU': tf.keras.metrics.MeanIoU})

model.summary()
```


**As observed from the results above, the model has a total of 28,306,817 parameters.**

1.Input: An image of size 5125123.

2.Intermediate Process: Features are extracted using ResNet50, followed by a simplified convolutional layer.

3.Output: A segmentation mask of size 5125121.
Purpose: Designed for binary segmentation tasks (e.g., extracting landfill boundaries from maps).
